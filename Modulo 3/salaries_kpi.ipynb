```python
import pandas as pd

df = pd.read_csv('Salaries.csv')

for colunas in df.columns:
    print(colunas)
```

```python
df = df.rename(columns={
    "Id": "Identificador",
    "EmployeeName": "NomeDoFuncionario",
    "JobTitle": "Cargo",
    "BasePay": "SalarioBase",
    "OvertimePay": "PagamentoHorasExtras",
    "OtherPay": "OutrosPagamentos",
    "Benefits": "Beneficios",
    "TotalPay": "PagamentoTotal",
    "TotalPayBenefits": "TotalPagamentoEBeneficios",
    "Year": "Ano",
    "Notes": "Notas",
    "Agency": "Agencia",
    "Status": "Status"
})

for colunas in df.columns:
    print(colunas)
```

```python
# Qual período, quais anos forma coletados ?
anos = df['Ano'].unique()
print(anos)

# Quantos e quais cargos foram registrados ?
cargos = sorted(df['Cargo'].unique())
for i in cargos:
    print(i)

# Qual a média de salários pagos ordenado do maior para o menor ?
mediaSalarios = df.groupby('Cargo')['TotalPagamentoEBeneficios'].mean().sort_values(ascending=False).head(10)
print(mediaSalarios)

# Qual o funcionário mais bem pago da lista ?
MaisBemPago = df.groupby('NomeDoFuncionario')['TotalPagamentoEBeneficios'].min().sort_values(ascending=False).head(1)
print(MaisBemPago)


# Quantos funcionários tem em cada cargo ?
funcionarios = df.groupby('Agencia')['Cargo'].value_counts().head(50)
print(funcionarios)

funcionarios.to_csv('relatório_cargos.csv')
print('relatório_cargos.csv')
```
