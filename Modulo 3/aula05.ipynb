```python
# caso precise
# %pip install pandas

import pandas as pd

# criar o DATA FRAME

df = pd.read_csv('tabela-fipe-historico-precos.csv')

display(df.head())

print(df.shape)

print(df.info())
```

```python
# Qual o carro mais caro de toda a série histórica?

# função "loc" localiza não apenas o campo, mas o registro inteiro.
maisCaro = df.loc[df['valor'].idxmax()]

print(maisCaro)
```

```python
# nunique - conta registros "ÚNICOS"
# quantas marcas já operaram no Brasil?
qtdMarcas = df['marca'].nunique()
print(f'Entre 1992 até 2022 {qtdMarcas} marcas operaram no Brasil')

qtdModelo = df['modelo'].nunique()
print(f'Entre 1992 até 2022 {qtdModelo} modelos operaram no Brasil')
```

```python
# unique - apresenta apenas os registros únicos
marcas = df['marca'].unique()
print(marcas)
```

```python
# criar ranking top 10 marcas mais valiosas
marcasTop10 = round(df.groupby('marca')['valor'].mean().sort_values(ascending=False).head(10))
print(marcasTop10)
```

```python
# Quais são os modelos dos 30 carros mais caros da série histórica
model30 = df.groupby(['marca','modelo'])['valor'].max().sort_values(ascending=False).head(30)
#print(model30)
print(model30.map(lambda x: f"R$ {x:,.2f}".replace(',','x'.replace('.',',').replace('x','.'))))

# explicação na próxima aula!
```
