# Criar um data frame (trazer os dados para a memória do python)


```python
import pandas as pd

df = pd.read_csv('dados.csv')
display(df)
```

```python
#head - imprime apenas uma amostra dos dados
print(df.head(2))

print('\n')

#info - apresenta um resumo do "dataset"
print(df.info())

print('\n')

#describe - soma a média. min, max... dos registros da tabela
print(df.describe())

print('\n')

# shape - mostra total de registros e colunas
print(df.shape)

print('\n')

#columns - identifica o titulo de cada coluna...
print(df.columns)
```

```python
marca = df['marca']
print(marca)

print('\n')

#consulta direto no print
print(df['modelo'])

print('\n')

#consultar mais que uma coluna
marca_modelo = df[['marca','modelo','cor']]
display(marca_modelo)
```

# Criar filtros

```python
# quero saber quantos carros não pagam mais ipva no meu estoque 
nipva = df[df['ano'] < 2006]
display(nipva)

print('\n')

# apresenta os que custam mais que 20 mil reais
ipva = df[df['valor'] > 20000]
display(ipva)

print('\n')

# apresente apenas os carros da chevrolet
CVT = df[df['marca'] == 'Chevrolet']
display(CVT)
```
