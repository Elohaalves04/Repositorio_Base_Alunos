## O que é um Seaborn?

É uma biblioteca para criar visualizações estatísticas em python com poucas linhas de código. 
OBS: Necessita do matplotlib.


```python
import matplotlib.pyplot as plt 
import seaborn as sns

# vamos usar um dataset de exemplo da própria biblioteca.
# Exemplos: Gorjetas, pinguins, titanic, diamantes, iris. 

gorjetas = sns.load_dataset('tips')
# print(gorjetas.head())
# print(gorjetas.shape)
# print(gorjetas.info())
# print(gorjetas.describe())
for i in gorjetas.columns: 
    print(i)
```

```python
# Gráfico de Disperção 

sns.scatterplot(
    data=gorjetas,
    x='total_bill',
    y='tip',
    hue='time'
)
plt.title('Relação entre o total da conta e as gorjetas')
plt.show()
```

```python
# Gráfico de barras 

sns.barplot(data=gorjetas,x='day',y='tip',hue='tip')
plt.show()
```

```python
# Gráfico de histograma 

sns.histplot(
    data=gorjetas,
    x='tip',
    kde=True
)

plt.title('Distribuição de Gorjetas')
plt.show()
```

```python
pinguins = sns.load_dataset('penguins')
print(pinguins.head())
```

```python
sns.set_theme(style='darkgrid')
sns.scatterplot(data=pinguins,x='body_mass_g',y='bill_depth_mm',hue='species')
plt.title('Dispersão entre peso e comprimento do pico por espécie')

# Importante salvar antes do plt.show()

plt.savefig(
    'grafico_dispesão.png',
    dpi=300,
)

plt.show()
```

```python
# Crie um gráfico de coluna entre peso e espécie

sns.barplot(
    data=pinguins,
    x='body_mass_g',
    y='species',
    hue='sex'
)
plt.show()
```

```python
sns.histplot(
    data=pinguins,
    x='body_mass_g',
    hue='species',
    kde=True
)

plt.show()
```

```python
sns.boxplot(
    data=pinguins,
    x='species',
    y='body_mass_g',
    hue='island',
)

```
