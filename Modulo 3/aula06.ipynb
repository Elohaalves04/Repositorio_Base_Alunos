## Aula 06 - Atividade Prática

### Pandas - Dataset - base_vendas 

### Você foi contratado como analista de dados e sua primeira atividade:

* Qual o faturamento da empresa? (vamos fazer juntos!)
* Quais produtos são vendidos?
* Qual o produto mais vendido? 
* Como foi o faturamente das vendas por gênero?
* Relacione as vendas por loja.

```python
# Criar a coluna faturamento
# Qtd Vendida * Preço Unitário

df['Faturamento'] = df['Qtd Vendida'] * df['Preço Unitário']
display(df[['Produto', 'Qtd Vendida', 'Preço Unitário', 'Faturamento' ]].head())

#OBS: A coluna faturamento existe apenas no data frame...
# No arquivo Base_vendas.xlsx a coluna não existe.
```

```python
# Identificar a coluna de produtos

produtos = df['Produto'].unique()

for i in produtos:
    print(i)
```

```python
# Identificar o produto mais vendido

pmv = df.groupby(['Produto'])['Qtd Vendida'].sum().sort_values(ascending=False).head(1)
print(pmv)
```

```python
# Faturamento das vendas por gênero

fgenero = df.groupby('Gênero')['Faturamento'].sum()
print(fgenero)
```

```python
# Relacione

fatLoja = df.groupby('Loja')['Faturamento'].sum()
print(fatLoja)
```
