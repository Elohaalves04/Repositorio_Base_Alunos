# Aula 07 - Gráficos Matplotlib

```python
import matplotlib.pyplot as plt

copasBrasil = {
    1958 : "Suécia",
    1962 : "Chile",
    1970 : "México",
    1994 : "Estados Unidos",
    2002 : "Coreia/Japão",
}

anos = copasBrasil.keys()
pais = copasBrasil.values()

plt.figure(figsize=(6,3))
plt.title('Anos em que o Brasil foi campeão mundial', color="#083064")
plt.plot(pais, anos, color="#3C8B3CFF") # gráfico de linhas, pais = x, anos = y
plt.xlabel('País Sede', color="#083064")
plt.ylabel('Ano ', color="#083064")

plt.show()
```

## Gráfico de Colunas


```python
meses = ['Janeiro','Fevereiro','Março','Abril','Maio','Junho']
vendas = [150, 285, 682, 450, 500, 550]

plt.figure(figsize=(6,3))
plt.bar(meses, vendas, color="#3ccacaff")
plt.title('Quantidade de Vendas - 1° Semestre 2026')

plt.show()
```

## Gráfico de Barras

```python
import pandas as pd
 
df = pd.read_excel('Base_vendas.xlsx')

df['Faturamento'] = df['Qtd Vendida'] * df['Preço Unitário']
display(df.head())

for i in df.columns:
    print(i)
```

```python
faturamentoLoja = df.groupby('Loja')['Faturamento'].sum().sort_values(ascending=False)
print(faturamentoLoja)

plt.figure(figsize=(15,4))
faturamentoLoja.plot(kind='barh')

plt.show()
```

## Gráfico de Pizza

```python
qtdGenero = df.groupby('Gênero')['Qtd Vendida'].sum()
print(qtdGenero)

plt.figure(figsize=(15,4))
pizza = qtdGenero.plot(kind='pie', colors=['red', 'blue'], autopct='%1.1f%%')
plt.title('Quantida vendida por gênero')

plt.show()
```
