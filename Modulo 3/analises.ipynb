```python
# Criando no notebook um dataframe.

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Compras.csv', sep=';')
display(df.head())
```

```python
# ver as colunas do dataset
print(df.columns)
```

## Criando 3 KPI's relevantes

```python
# Transformando em data

df['Data'] = pd.to_datetime(df['Data'], format='%d/%m/%Y')
print(df.info())
```

```python
# Qual a data que teve mais PRODUTOS VENDIDOS?
dataMaisVendas = df.groupby(['Data'])['Quantidade'].sum().sort_values(ascending=False).head(1)
print(dataMaisVendas)
```

```python
# Qual é a maior venda financeira?
maisCaro = df.loc[df['ValorFinal'].idxmax()]
print(maisCaro)
```

```python
# Qual é o produto mais vendido?
maisVendido = df.groupby(['Produto'])['Quantidade'].sum().sort_values(ascending=False).head(1)
print(maisVendido)
```

# Mais opções de KPI's (Feitas pelo Professor)

```python
df['Região'] = df['Fornecedor'].str.split().str[-1]
display(df.head(20))
```

```python
top5_produtos = df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False).head(5)
print(top5_produtos)
```

```python
top1_Fornecedor = df.groupby('Fornecedor')['Quantidade'].sum().idxmax()
print(f'O fornecedor que mais vendeu foi {top1_Fornecedor}')
```
